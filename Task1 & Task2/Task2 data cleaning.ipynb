```python
import pandas as pd
```

```python
data = pd.read_excel('sample physician information.xlsx')
final_df = pd.DataFrame()

for i in data.index:
    
    # for name
    strr = data['Fulll Information'][i]
    str_spl_name = strr.split('(')[0]
    print(str_spl_name)
    
    #for CPSO no
    strr2 = data['Fulll Information'][i]
    cpso = strr2.split('CPSO#:')[1]
    str_spl_cpso = cpso.split(')')[0]
    print(str_spl_cpso)
    
    # for phone no
    str3 = data['Fulll Information'][i]
    parse_str3 = str3.split('Phone:')[1]
    str_stripped_phone = parse_str3[0:14]
    str_spl_phone = str_stripped_phone.strip()
    print(str_spl_phone)
    
    # for aos
    str4 = data['Fulll Information'][i]
    try:
        parse_str3 = str4.split('Area(s) of Specialization:')[1]
    except:
        parse_str3 = 'nan'
    str_spl_aos = parse_str3.strip()
    print(str_spl_aos)
    
    data_in_list = [[str_spl_cpso, str_spl_name, str_spl_phone, str_spl_aos]]
 
    # using append
    final_df = final_df.append(pd.DataFrame( data_in_list,
               columns=[ 'CPSO #', 'Name', 'Phone no', 'Area of Specialization']),
               ignore_index = True)
    
    
# -	Get Area of Specialization wise count of physicians
group_on_aos = final_df.groupby(['Area of Specialization']) 
print(group_on_aos.count())


#finally saving the file.    
final_df.to_excel("Task2 output.xlsx") 
print("file saved successfully...")
```
